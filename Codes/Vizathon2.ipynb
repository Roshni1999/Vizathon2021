```python
#removing the columns having most null values
covid_vaccine_statewise = covid_vaccine_statewise.drop(['Total Sputnik V Administered','AEFI', '18-45 years (Age)','45-60 years (Age)','60+ years (Age)'], axis=1)
```

```python
#Genderwise pie-chart for vaccination
import plotly.express as px
import matplotlib.pyplot as plt
male = covid_vaccine_statewise["Male(Individuals Vaccinated)"].sum() 
female = covid_vaccine_statewise["Female(Individuals Vaccinated)"].sum()  
trans = covid_vaccine_statewise["Transgender(Individuals Vaccinated)"].sum()
px.pie(names=["Male Vaccinated","Female Vaccinated","Trans Gender"],values=[male,female,trans],title="Male and Female Vaccinated ratio for Covid19")




```

```python
v=covid_vaccine_statewise[covid_vaccine_statewise.State!='India']
print(v.index)
```

```python
v.head(5)
```

```python
v.rename(columns = {"Total Individuals Vaccinated": "Total"}, 
          inplace = True)
```

```python
States_grouped=v.groupby('State')['Total'].sum().to_frame('Total')
States_grouped=States_grouped.sort_values('Total',ascending=False).head(5)
States_grouped
```

```python
fig=plt.figure(figsize=(10,11))
plt.title("Top 5 Vaccinated states", size=22)
plt.xlabel("states")
plt.ylabel("Vaccination number")
x=sns.barplot(data=States_grouped.iloc[:10],y=States_grouped.Total,x=States_grouped.index, linewidth=2).set(xlabel='States', ylabel='Number of confirmed cases')
plt.show()
```
